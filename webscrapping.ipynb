```python
#Step 2: Send Request to Amazon.com
import requests

url = "https://www.amazon.com/s?k=laptops"
response = requests.get(url)

print(response.status_code)
```

pass HTML with BeautifulSoup
The BeautifulSoup function takes two arguments: the HTML content to parse and the parser to use. We use the html.parser parser to parse the HTML content.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
```

Step 4: Extract Product Details
code loops through each search result on the page and extracts the name, price, and rating of each laptop. We use the find_all function to find all the div elements with the attribute data-component-type set to 's-search-result'. We then use the find function to find the name, price, and rating elements within each search result.

```python
products = []

for result in soup.find_all('div', {'data-component-type': 's-search-result'}):
    product = {
        'name': result.find('h2').text.strip(),
        'price': result.find('span', {'class': 'a-offscreen'}).text.strip(),
        'rating': result.find('span', {'class': 'a-icon-alt'}).text.strip()
    }
    products.append(product)
```

Step 5: Display Data

```python
import pandas as pd

df = pd.DataFrame(products)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page to scrape
url = "https://en.wikipedia.org/wiki/Web_scraping"

# Make a request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the title of the Wikipedia page
title = soup.find("h1", {"id": "firstHeading"}).text
print("Title: ", title)

# Find the introduction section of the page
intro = soup.find("div", {"class": "mw-parser-output"}).p.text
print("Introduction: ", intro)



# Find all the links in the page
links = soup.find_all("a")
for link in links:
    print(link.get("href"))
```

Beautiful Soup
Beautiful Soup is a Python library that works with a parser to extract data from HTML and can turn even invalid markup into a parse tree. However, this library is only designed for parsing and cannot request data from web servers in the form of HTML documents/files. For this reason, it is mostly used alongside the Python Requests Library.
 Note that Beautiful Soup makes it easy to query and navigate the HTML, but still requires a parser. Code demonstrates the use of the html.parser module, which is part of the Python Standard Library.

Part 1  Get the HTML using Requests

```python
import requests
url = 'https://oxylabs.io/blog'
response = requests.get(url)
```

```python
print(response)
```

Part 2  Find the element



```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
```

Due to its simple ways of navigating, searching and modifying the parse tree, Beautiful Soup is ideal and saves developers hours of work. For example, to print all the blog titles from this page, the findAll() method can be used. On this page, all the blog titles are in h3 , h4 , h5 and elements. This information can be supplied to the findAll method as follows:

```python
blog_titles = soup.findAll(['h3', 'h4', 'h5'])
for title in blog_titles:
    print(title.text)
```

BeautifulSoup also makes it easy to work with CSS selectors. If a developer knows a CSS selector, there is no need to learn find() or find_all() methods. The following is the same example but uses CSS selectors to remove the "Most popular articles" title that's in h5 from being included in the blog_titles object:
While broken-HTML parsing is one of the main features of this library, it also offers numerous functions, including the fact that it can detect page encoding further increasing the accuracy of the data extracted from the HTML file.

What is more, it can be easily configured, with just a few lines of code, to extract any custom publicly available data or to identify specific data types.

```python
blog_titles = soup.select('h3, h4, h5:not(:-soup-contains("Most popular articles"))')
for title in blog_titles:
    print(title.text)
```

This library contains a module html to work with HTML. However, the lxml library needs the HTML string first. This HTML string can be retrieved using the Requests library as discussed in the previous section. Once the HTML is available, the tree can be built using the fromstring method as follows:

```python
# After response = requests.get()
from lxml import html
tree = html.fromstring(response.text)
```


//h3 | //h4 | //h5[not(contains(text(), "Most popular articles"))]
This XPath can be given to the tree.xpath() function. This will return all the elements matching this XPath. Notice the text() function in the XPath. This will extract the text within the h3, h4, and h5 elements.

```python
blog_titles = tree.xpath('//h3 | //h4 | //h5[not(contains(text(), "Most popular articles"))]')
for title in blog_titles:
    print(title.text)
```

```python
import csv
from lxml import html
import requests

# Make a GET request to the website
url = 'https://oxylabs.io/blog'
response = requests.get(url)

# Create an HTML tree from the response content
tree = html.fromstring(response.content)

# Extract the blog titles
blog_titles = tree.xpath('//h3 | //h4 | //h5[not(contains(text(), "Most popular articles"))]')

# Write the titles to a CSV file
with open('titles.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for title in blog_titles:
        writer.writerow([title.text])
```
