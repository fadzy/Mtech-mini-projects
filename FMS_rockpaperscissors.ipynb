```python
import random
import matplotlib.pyplot as plt

# Define the possible moves in rock paper scissors
moves = ["rock", "paper", "scissors"]

# Define the transition dictionaries for the FSMs
transition_dicts = [
    {"rock": "paper", "paper": "scissors", "scissors": "rock"},
    {"rock": "scissors", "paper": "rock", "scissors": "paper"},
    {"rock": "rock", "paper": "paper", "scissors": "scissors"}
    # Add more transition dictionaries as needed
]

# Define the size of the population
population_size = 10

# Generate a population of FSMs with random transition dictionaries
population = [random.choice(transition_dicts) for i in range(population_size)]

# Define the number of rounds to play for each evaluation
num_rounds = 10

# Define a function to play a round of rock paper scissors
def play_round(move, other_move):
    if move == other_move:
        return 0  # Tie
    elif (move == "rock" and other_move == "scissors") or \
         (move == "paper" and other_move == "rock") or \
         (move == "scissors" and other_move == "paper"):
        return 1  # Win
    else:
        return -1  # Loss

# Define a function to evaluate the fitness of an FSM
def evaluate_fsm(fsm):
    total_score = 0
    for i in range(num_rounds):
        my_move = random.choice(moves)
        other_move = random.choice(moves)
        my_next_move = fsm[my_move]
        score = play_round(my_next_move, other_move)
        total_score += score
    return total_score

# Define the number of generations to run
num_generations = 50

# Define the probability of mutation
mutation_prob = 0.1

# Define lists to store the maximum and average fitnesses for each generation
max_fitnesses = []
avg_fitnesses = []

# Run the evolutionary algorithm for the specified number of generations
for generation in range(num_generations):

    # Evaluate the fitness of each FSM in the population
    fitness_scores = [evaluate_fsm(fsm) for fsm in population]

    # Print the maximum and average fitness for this generation
    max_fitness = max(fitness_scores)
    avg_fitness = sum(fitness_scores) / population_size
    print(f"Generation {generation}: Max Fitness = {max_fitness}, Avg Fitness = {avg_fitness}")
    max_fitnesses.append(max_fitness)
    avg_fitnesses.append(avg_fitness)

    # Select the fittest FSMs to reproduce
    sorted_indices = sorted(range(len(fitness_scores)), key=lambda k: fitness_scores[k], reverse=True)
    fittest_population = [population[i] for i in sorted_indices[:int(population_size/2)]]

    # Reproduce to create a new population
    new_population = []
    for i in range(population_size):
        parent1 = random.choice(fittest_population)
        parent2 = random.choice(fittest_population)
        child_dict = {}
        for move in moves:
            if random.random() < 0.5:
                child_dict[move] = parent1[move]
            else:
                child_dict[move] = parent2[move]
            if random.random() < mutation_prob:
                child_dict[move] = random.choice(moves)
        new_population
```

```python
#Plot the final maximum fitness and the corresponding FSM
best_index = fitness_scores.index(max(fitness_scores))
best_fsm = population[best_index]
print(f"Best FSM: {best_fsm}")
print(f"Fitness score: {max(fitness_scores)}")

#Plot the maximum and average fitnesses for each generation
plt.plot(range(num_generations), max_fitnesses, label="Max Fitness")
plt.plot(range(num_generations), avg_fitnesses, label="Average Fitness")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.legend()
plt.show()
```
