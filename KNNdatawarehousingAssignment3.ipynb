```python
import numpy as np

def euclidean_distance(x1, x2):

    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
from collections import Counter
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = []
        for i in range(len(X)):
            distances = [euclidean_distance(X[i], x_train) for x_train in self.X_train]
            k_idx = np.argsort(distances)[:self.k]
            k_neighbor_labels = [self.y_train[idx] for idx in k_idx]
            most_common = Counter(k_neighbor_labels).most_common(1)
            y_pred.append(most_common[0][0])
        return np.array(y_pred)



# Example usage:
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])
y = np.array([0, 0, 1, 1, 0, 1])  # binary class labels
knn = KNN(k=3)
knn.fit(X, y)
X_test = np.array([4, 4])#, [4, 5], [7, 6]])
print(knn.predict(X_test))
# Create KNN classifier object
knn = KNN(k=3)

# Define input data and class labels
#X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])
#y = np.array([0, 0, 1, 1, 0, 1])

# Train the KNN model
#knn.fit(X, y)
```

```python

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
# Train the KNN model
knn.fit(X, y)
# Plot the original data points
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k')
plt.title('Input data points')
plt.show()

# Generate a mesh grid of points to classify
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Classify each point on the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the classified mesh grid
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
# Plot the test point as a red dot
plt.scatter(X_test[0], X_test[1], marker='x', color='red')

# Plot the original data points again
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('KNN classification')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Example data
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])
y = np.array([0, 0, 1, 1, 0, 1])  # binary class labels

# Create a kNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the data
knn.fit(X, y)

# Example test data
X_test = np.array([[4, 4]])

# Predict the class of the test data
y_pred = knn.predict(X_test)

print(y_pred)
```
